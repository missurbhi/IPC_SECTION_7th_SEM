```python
import os
from openai import openai
```

```python
client=OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
response=client.chat.completions.create(
    model="gpt-40",
    messages=[
        {
            "role" :"user",
            "content":""" please summasize in 50 words.
            Openai is a leading ar research organization." 
        }
    ]
)
```

```python
print(response)
```
